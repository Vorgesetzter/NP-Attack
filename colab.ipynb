## 3. Results

```
# Find and download the output file
import glob
from google.colab import files

output_files = glob.glob('logs/**/*.wav', recursive=True) + glob.glob('*.wav')
output_files = [f for f in output_files if 'predictor' in f]

if output_files:
    print(f"Downloading: {output_files[0]}")
    files.download(output_files[0])
else:
    print("No output file found")
```

```
# Listen to results
import IPython.display as ipd

print("Original:")
ipd.display(ipd.Audio("ground_truth.wav"))

if output_files:
    print("\nAdversarial:")
    ipd.display(ipd.Audio(output_files[0]))
```
